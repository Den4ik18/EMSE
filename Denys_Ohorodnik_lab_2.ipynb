# Ініціалізація датасету


```
# підключимо бібліотеку pandas для обробки даних (вона вже встановлена в Colab)
import pandas as pd

# зазначимо url звідки ми візьмемо дані
url = "http://www.sharecsv.com/dl/70792028b87f9e5072133ce87c32a8a7/googleplaystore.csv"

# створимо pandas Dataframe з csv-файлу, який ми отримаємо по вищезазначеному url
# Dataframe надасть нам необхідні можливості для обробки даних
df = pd.read_csv(url)

# очистимо dataframe, підготуємо його до бажаного вигляду
#df.set_index('App', inplace=True)  # встановимо колонку з іменем '№' як індекс dataframe'у

# конвертуємо колонку "Reviews" у числові значення.
#df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')  

# відобразимо перші 5 записів у датафреймі
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df.head()
```

## Bar Plots

```
import matplotlib.pyplot as plt
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

#x2 = [2, 4, 6, 8, 10]
#y2 = [5, 6, 2, 6, 2]
x = df.Rating
#dfList1 = df['Rating'].tolist()
#y = df.Reviews

# Colors: https://matplotlib.org/api/colors_api.html

plt.bar(range(0,len(x)),x, label="Rating", color='b')

plt.plot()

plt.xlabel("rating length")
plt.ylabel("rating height")
plt.title("Rating Chart ")
plt.legend()
plt.show()
```

##Deplot

```
# library & dataset
import seaborn as sns
#df = sns.load_dataset('googleplaystore.csv')

 
# density plot with shade
sns.kdeplot(df['Reviews'], shade=True)
sns.plt.show()
```

##Pie chart

```
import matplotlib.pyplot as plt

labels = '159', '967', '87510'
sections = [len(df[df.Reviews == 159].index),
            len(df[df.Reviews == 967].index),
            len(df[df.Reviews == 87510].index)]
colors = ['#ed3a01', '#21e12c', '#0095aa']

plt.pie(sections, labels=labels, colors=colors,
        startangle=90,
        autopct = '%1.2f%%')

plt.axis('equal')

plt.show()
```

## Circle chart

```
import matplotlib.pyplot as plt
 
# create data
#size_of_groups=[12,11,3,30]

dfList1 = [len(df[df.Reviews <= 159.0]),
            len(df[df.Reviews >= 967.0]),
            len(df[df.Reviews > 87510.0])] 
# Create a pieplot
plt.pie(dfList1)
#plt.show()
 
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
 
plt.show()
#зелений - значення більше 967
#синій - значення більше 87510
#червоний - значення менше 159 
```

## Hist chart

```
import numpy as np
import matplotlib.pyplot as plt

h = df.Rating 
plt.hist(h, range = (1,5))
plt.title("Rating")
plt.xticks(np.arange(1,5, step=0.5)+0.5, range(1,5))
plt.show()
```
